#### Chi - square test of independence

```python
type_of_cab_crosstab = pd.crosstab(df['Type_of_Cab'],df['Surge_Pricing_Type'],margins = True)
type_of_cab_crosstab
```

```python
conf_lifestyleIndex_crosstab = pd.crosstab(df['Confidence_Life_Style_Index'],df['Surge_Pricing_Type'],margins = True)
conf_lifestyleIndex_crosstab
```

```python
destination_crosstab = pd.crosstab(df['Destination_Type'],df['Surge_Pricing_Type'],margins = True)
destination_crosstab
```

```python
gender_crosstab = pd.crosstab(df['Gender'],df['Surge_Pricing_Type'],margins = True)
gender_crosstab
```

```python
from scipy import stats
```

```python

def check_categorical_dependency(crosstab_table):
    stat,p,dof,expected = stats.chi2_contingency(crosstab_table)
    print("Chi-Square static value = {}".format(stat))
    print("p-value = {}".format(p))
    
    alpha = 0.05
    if p < alpha:
        print("Dependent(Reject H0)")
    else:
        print("Independent(Accept H0)")
```

```python
check_categorical_dependency(destination_crosstab) 
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,train_size = 0.7,random_state = 0)
x_train.shape
```

#### Simple Gradient boosted decision tree

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
clf = GradientBoostingClassifier(n_estimators=5000, learning_rate=0.01, max_depth=3, random_state=0)
clf.fit(x_train, y_train)
```

```python
yhat_boosted = clf.predict(x_test)
accuracy_score(yhat_boosted,y_test)
```

```python
yhat_boosted_submit = clf.predict(df_test)
```

```python
id = np.asanyarray(df_test.index)
df_submit = pd.DataFrame({'Trip_ID':id,'Surge_Pricing_Type':yhat_boosted_submit})
df_submit.head()
```

```python
compression_opts = dict(method='zip',
                        archive_name='simple_gradient_boosted.csv')  
df_submit.to_csv('out1.zip', index=False,
          compression=compression_opts)
```
